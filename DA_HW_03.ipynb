```python
def primo (numero):
    if numero < 2:
        return 'Escribe un numero mayor a 1'
    for n in range (2,numero):
        if numero % n == 0:
            return False
    return True
        
```

```python
def es_positivo(numero):
    if numero >= 0:
        return True
    else:
        return False
```

```python
def es_positivo(numero):
    try:
        if float(numero) > 0:
            return True
        else:
            return False
    except ValueError:
        print("Error: El parámetro no es un número.")
        return False
```

```python
def division_segura(dividendo, divisor):
    try:
        resultado = float(dividendo) / float(divisor)
        return resultado
    except (ValueError, TypeError):
        print("Error: Los parámetros deben ser números.")
        return None
    except ZeroDivisionError:
        print("Error: División por cero.")
        return None
```

6) Crea una función que al recibir una lista de números, devuelva el que más se repite y cuántas veces lo hace. Si hay más de un "más repetido", que devuelva cualquiera

```python
def numero_que_mas_se_repite(lista):
    lista_de_numeros=[]
    lista_de_repeticiones=[]
    if len(lista)==0:
        return None
    for numero in lista:
        if numero in lista_de_numeros:
            ind=lista_de_numeros.index(numero)
            lista_de_repeticiones[ind] += 1
        else:
            lista_de_numeros.append(numero)
            lista_de_repeticiones.append(1)
    variable_numero_mayor=lista_de_numeros[0]
    variable_cantidad_repeticiones=lista_de_repeticiones[0]
    for ind, numero in enumerate(lista_de_numeros):
        if lista_de_repeticiones[ind] > variable_cantidad_repeticiones:
            variable_numero_mayor=lista_de_numeros[ind]
            variable_cantidad_repeticiones=lista_de_repeticiones[ind]
    return variable_numero_mayor, variable_cantidad_repeticiones
```

```python
def numero_que_mas_se_repite2(lista):
    
    lista_de_repeticiones=[]
    if len(lista)==0:
        return None
    else:
        lista_de_numeros=list(set(lista))
        for numero in lista_de_numeros:
            lista_de_repeticiones.append(lista.count(numero))
    variable_numero_mayor=lista_de_numeros[0]
    variable_cantidad_repeticiones=lista_de_repeticiones[0]
    for ind, numero in enumerate(lista_de_numeros):
        if lista_de_repeticiones[ind] > variable_cantidad_repeticiones:
            variable_numero_mayor=lista_de_numeros[ind]
            variable_cantidad_repeticiones=lista_de_repeticiones[ind]
    return variable_numero_mayor, variable_cantidad_repeticiones
```

7) Crea una función que convierta entre grados Celsius, Farenheit y Kelvin<br>
Fórmula 1	: (°C × 9/5) + 32 = °F<br>
Fórmula 2	: °C + 273.15 = °K<br>
Debe recibir 3 parámetros: el valor, la medida de orígen y la medida de destino


```python
def convertidor_temperatura(valor,origen,destino):
    '''Funcion que realiza la conversion de diferentes unidades de grados
    Parametros
    -valor: el valor numerico a convertir
    -origen: la unidad de origen C: Celsius, F: Farenheit, K: Kelvin
    -destino: la unidad de destino: C: Celsius, F: Farenheit, K: Kelvin'''
    if origen == 'C':
        if destino == 'C':
            valor_destino = valor
        elif destino == 'F':
            valor_destino = (valor*(9/5))+32
        elif destino == 'K':
            valor_destino = valor + 273.15
        else:
            print('Verifique parametro de destino')
    elif origen == 'F':
        if destino == 'F':
            valor_destino = valor
        elif destino == 'C':
            valor_destino = (valor-32)*(5/9)
        elif destino == 'K':
            valor_destino = ((valor-32)*(5/9)) + 273.15
        else:
            print('Verifique parametro de destino')
    elif origen == 'K':
        if destino == 'K':
            valor_destino = valor
        elif destino == 'F':
            valor_destino = ((valor - 273.15)*(9/5))+32
        elif destino == 'C':
            valor_destino = valor - 273.15
        else:
            print('Verifique parametro de destino')
    else:
        print('Verifica el parametro de origen')
    return valor_destino
```

9) Arma una función que devuelva el factorial de un número. Tener en cuenta que el usuario puede equivocarse y enviar de parámetro un número no entero o negativo

```python
def factorial(numero):
    '''Calcula el factorial de un numero'''
    if type(numero) != int:
        return 'Solo se calcula el factorial de un numero entero'
    elif numero < 0:
        return 'No es posible el factorial de un numero negativo'
    elif numero == 0:
        return 1
    else:
        numero = numero * factorial(numero-1)
    return numero
```
